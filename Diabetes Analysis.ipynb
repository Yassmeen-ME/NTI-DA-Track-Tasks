```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
df_diabetes= pd.read_csv("D:/Yassmeen/NTI-Data Analysis/S6 Matplotlib & Seaborn/Task 6/diabetes.csv")
```

```python
#No. of Diabetic vs Not Diabetic (count of Outcome)
df_diabetes['Outcome'].value_counts()
```

```python
#No of pregnancies
df_diabetes['Pregnancies'].value_counts()
```

```python
#No of pregnancies based on outcome
df_diabetes.groupby('Outcome')['Pregnancies'].value_counts()
```

```python
#what's the correlation between no. of pregrancies and being diabetic
df_diabetes['Pregnancies'].corr(df_diabetes['Outcome'])  
#negative correlation
```

```python
#correlation (using heatmap)
#df_diabetes.corr()
plt.title('Correlation Heatmap')
sns.heatmap(data=df_diabetes.corr(), annot=True)
plt.show()
```

```python
#No. of Diabetic vs Not Diabetic (count of Outcome)
plt.title('No of Healthy vs Diabetic')
plt.pie(df_diabetes['Outcome'].value_counts(), explode=(0, 0.05), shadow=True, labels=['Healthy','Diabetic'], autopct='%1.2f%%', startangle=180)
```

```python
#Age Distribution
df_diabetes.Age.plot(kind='hist')
plt.title('Age Distribution')
plt.show()
```

```python
#Relation between Age & Glucose based on outcome
sns.scatterplot(data=df_diabetes, x='Age' , y='Glucose', hue='Outcome')
plt.title('Relation between Age & Glucose based on Outcome')
plt.show()
```

```python
#Relation between Age & BMI based on outcome
sns.scatterplot(data=df_diabetes, x='Age' , y='BMI', hue='Outcome')
plt.title('Relation between Age & BMI based on Outcome')
plt.show()
```

```python
#FEATURES dISTRIBUTION
df_diabetes.hist(figsize=(14,8))
plt.suptitle('Features Distribution')
plt.show()
```

```python
#Relation between Pregnancies & Outcome
PO_relation= df_diabetes.groupby('Outcome')['Pregnancies'].mean().reset_index()
sns.barplot(data=PO_relation,x='Outcome', y='Pregnancies')
plt.title('Average Pregnancies by Diabetes Outcome')
plt.show()
```

```python
#relation between blood pressure & Outcome
sns.scatterplot(data=df_diabetes, y='Outcome', x='BloodPressure')
plt.title('Relation between Blood Pressure & Outcome')
plt.show()
```

```python
#Average of Age based on Oucome
plt.title('Average of Age based on Oucome')
df_diabetes_filtered = df_diabetes.groupby('Outcome')['Age'].mean().reset_index()
sns.barplot(y=df_diabetes_filtered['Age'], x=df_diabetes_filtered['Outcome'])
plt.show()
```
