```python
#worst Selling course based on num_subscribers 
#show worst 5 & best 5
df_Udemy = df_Udemy[df_Udemy['num_lectures'] != 0]
```

```python
#Average price for each Subject      new1
#first relace free with 0 , change col data type to float, then check
df_Udemy['price'] = df_Udemy['price'].replace('Free', 0).astype(float)
df_Udemy
```

```python
#Average price for each Subject      new1
Avg_price= df_Udemy.groupby('subject')['price'].mean()
Avg_price= Avg_price.round(2)
print(Avg_price)
```

```python
#drop courses with 0 lectures then check
df_Udemy = df_Udemy[df_Udemy['num_lectures'] != 0]
df_Udemy.sort_values('num_lectures')
```

```python
#courses in Graphic design price >= 100 new2
df_Udemy[(df_Udemy['subject'] == 'Graphic Design') & (df_Udemy['price'] >=100)]
```

```python
#courses that contains word python
df_Udemy[df_Udemy['course_title'].str.contains('Python')]
```

```python
#published courses in 2017 
# first:change published_timestamp to datetime , then check
df_Udemy.published_timestamp = pd.to_datetime(df_Udemy.published_timestamp)
df_Udemy.dtypes
```

```python
#published courses in 2017 
#second: create col: year from published_timestamp , then check
df_Udemy['year'] = df_Udemy.published_timestamp.dt.year
df_Udemy
```

```python
#published courses in 2017 
#third: etract 2017 only
df_Udemy[df_Udemy.year == 2017]
```

```python
#max num_subscribers in each level
df_Udemy.groupby('level')[['num_subscribers']].max()
```

```python
#Max in each level (Numerical only)
df_Udemy.groupby('level').max(numeric_only=True)
```

```python
#courses that belongs to Web Development & Business Finance	
df_Udemy[df_Udemy.subject.isin(['Web Development' , 'Business Finance'])]
```

```python
#Correlatiob between features
df_Udemy_numerical= df_Udemy.select_dtypes(exclude=np.object_)
plt.figure(figsize=(6,6))
sns.heatmap(df_Udemy_numerical.corr(), annot=True)
```

```python
#Visual: No. of subscribers in each course
plt.figure(figsize=(15,5))
plt.title('No. of Subscribers per course', fontsize=20 )
plt.bar(df_Udemy.index, df_Udemy.num_subscribers , color='r')
plt.xlabel('Index', fontsize=14)
plt.ylabel('num_subscribers', fontsize=14)
```

```python
#Max duration hour in each level
df_Udemy_hours= df_Udemy.groupby('level')[['content_duration']].max()
df_Udemy_hours
```

```python
#Visual: Max duration hour in each level
#plt.figure(figsize=(15,5))
plt.plot(df_Udemy_hours.index, df_Udemy_hours['content_duration'])
plt.title('Max duration hour per level', fontsize=20 )
plt.xlabel('level', fontsize=14)
plt.ylabel('Duration', fontsize=14)
plt.show
```

```python
#courses prices
df_Udemy_prices= df_Udemy.groupby('course_id')[['price']].max()
df_Udemy_prices
```

```python
#Visual:courses prices
plt.figure(figsize=(15,5))
plt.plot(df_Udemy_prices.index, df_Udemy_prices['price'])
plt.title('courses price', fontsize=20 )
plt.xlabel('courses', fontsize=14)
plt.ylabel('price', fontsize=14)
plt.show
```

```python

#No. of lectures per course
plt.figure(figsize=(15,5))
plt.title('No. of lectures per course', fontsize= 20)
sns.distplot(df_Udemy['num_lectures'], bins=500)
plt.xlim(0,200)  #to focus on particular area
plt.show
```

```python
#No of subscribers in paid vs free courses
plt.figure(figsize=(15,5))
plt.title('No of subscribers in paid vs free courses')
sns.scatterplot(x='num_subscribers', y='course_id' , data=df_Udemy , hue='is_paid')
plt.show()
```

```python
#Total price over years
plt.figure(figsize=(6,6))
price_per_year = df_Udemy.groupby('year')['price'].sum()
plt.plot(price_per_year.index, price_per_year.values, marker='o')
plt.title('Total Course Price Over Years')
plt.show()
```

```python
#distribution of published corses over months
df_Udemy['month']=df_Udemy['published_timestamp'].dt.month_name()
plt.figure(figsize=(12,6))
sns.countplot(data=df_Udemy , x='month')
plt.title('distribution of published corses over months')
plt.show()
```

```python
#highest 5 courses with highest subscribers
plt.figure(figsize=(16,6))
top5= df_Udemy.nlargest(5, 'num_subscribers')
sns.catplot(data=top5 , kind='bar', x='course_title' , y='num_subscribers')
plt.title('Top 5 Courses by Number of Subscribers')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#highest 5 courses based on reviews
z=df_Udemy.loc[df_Udemy.num_reviews.sort_values(ascending=False).head().index]
sns.catplot(data=z, kind='bar', x='course_title', y='num_reviews', height=5, aspect=2)
plt.title('highest 5 courses based on reviews')
plt.xticks(rotation=45, ha='right')
```

```python
#Top 10 courses based on price
top10_by_price = df_Udemy.sort_values(by='price', ascending=False).head(10) 
#it will rturn 10 random values where price is the highest(200)
sns.catplot(data=top10_by_price, kind='bar', x='price', y='course_title' , height=7, aspect=2.2)
plt.title('Top 10 courses based on price', fontsize=20)
plt.show()
```

```python
#top 10 courses based on num_lectures
top10_by_lect_num = df_Udemy.sort_values(by='num_lectures', ascending=False).head(10)
sns.catplot(data= top10_by_lect_num, kind='bar', x='num_lectures', y='course_title' , height=7, aspect=2.2)
plt.title('Top 10 courses based on No. of Lectures', fontsize=20)
plt.show()
```
